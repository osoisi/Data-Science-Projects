```python
import pandas as pd
```

```python
df= pd.read_csv(r'HealthcareDiabetes.csv')
```

```python
df.head(20)
```

```python
df.info()
```

```python
df.fillna('None', inplace=True)
```

```python
symptom_cols = [c for c in df.columns if 'Symptom' in c]
```

```python
symptom_cols
```

```python
all_symptoms = set()
for col in symptom_cols:
    all_symptoms.update(df[col].unique())
```

```python
all_symptoms
```

```python
symptom_to_int = {symptom: i for i, symptom in enumerate(all_symptoms)}
```

```python
symptom_to_int
```

```python
for col in symptom_cols:
    df[col] = df[col].map(symptom_to_int)
```
